```
from tensorflow.keras.applications.vgg16 import VGG16,preprocess_input,decode_predictions 

from tensorflow.keras.preprocessing.image import load_img,img_to_array
```

```
model = VGG16()
model.summary()
```

```
import os

import matplotlib.pyplot as plt
import matplotlib.image as mpimg

for file in os.listdir("/content/drive/My Drive/Colab Notebooks/KGP Talkie DL/some"):
    print(file)
    full_path = "/content/drive/My Drive/Colab Notebooks/KGP Talkie DL/some/" + file

    # Printing the Images
    image = load_img(full_path,target_size=(224,224))
    plt.imshow(image)
    plt.show()

    # Pre-pro
    image = img_to_array(image)
    image = image.reshape((1,image.shape[0],image.shape[1],image.shape[2]))
    image = preprocess_input(image)
    y_pred = model.predict(image)

    # Top Predicted Labels
    label = decode_predictions(y_pred,top = 1)
    # label = decode_predictions(y_pred,top = 2)

    # Separate code for showing the images
    # img=mpimg.imread(image)
    # imgplot = plt.imshow(img)
    # plt.show()


    print(label)
    print()
```
