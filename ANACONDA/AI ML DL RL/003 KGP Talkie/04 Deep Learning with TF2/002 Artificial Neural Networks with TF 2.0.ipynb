```
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.model_selection import train_test_split
```

```
dataset = pd.read_csv(r"/content/drive/My Drive/Colab Notebooks/Data Folder/Churn_Modelling.csv")
```

```
dataset.head()
```

```
X = dataset.drop(labels=['CustomerId','Surname','RowNumber','Exited'],axis=1)
y = dataset['Exited']
```

```
X.head()
```

```
y.head()
```

```
from sklearn.preprocessing import LabelEncoder
```

```
# Converting Categorical Data to Numerical Data
label1 = LabelEncoder()
X['Geography'] = label1.fit_transform(X['Geography'])
X.head()
```

```
# Converting Categorical Data to Numerical Data
label2 = LabelEncoder()
X['Gender'] = label2.fit_transform(X['Gender'])
X.head()
```

```
X = pd.get_dummies(X,drop_first=True,columns=['Geography']) # no Geography_3 (redundant)
X.head()
```

# Feature Standardization

```
from sklearn.preprocessing import StandardScaler
```

```
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2 ,random_state = 0 , stratify = y )
```

```
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```
X_train
```

# Build ANN

```
model = Sequential()
model.add(Dense(X.shape[1],activation="relu",input_dim = X.shape[1]))
model.add(Dense(128,activation="relu"))
model.add(Dense(1,activation="sigmoid"))
```

```
X.shape[1]
```

```
model.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
```

```
# converting y_train from pandas series to numpy array
# Stochastic gradient Descent becomes Batch gradient descent
model.fit(X_train,y_train.to_numpy(),batch_size=10,epochs=10,verbose =1)
```

```
# Trying with more epochs
model = Sequential()
model.add(Dense(X.shape[1],activation="relu",input_dim = X.shape[1]))
model.add(Dense(128,activation="relu"))
model.add(Dense(1,activation="sigmoid"))
model.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
model.fit(X_train,y_train.to_numpy(),batch_size=10,epochs=20,verbose =1)  
```

```
y_pred = model.predict_classes(X_test)
```

```
y_pred
```

```
y_test
```

```
model.evaluate(X_test,y_test.to_numpy())
```

```
from sklearn.metrics import confusion_matrix,accuracy_score
```

```
confusion_matrix(y_test,y_pred)
```

```
accuracy_score(y_test,y_pred)
```
