```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn import datasets,metrics
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
cancer = datasets.load_breast_cancer()
cancer.keys()
# cancer
```

```python
cancer.target_names
```

```python
cancer.feature_names
```

```python
cancer.feature_names.shape
```

```python
X = cancer.data
y = cancer.target
X.shape,y.shape
```

```python
X[0:3]
```

```python
y
```

# Standardization

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

# Split the data and the model

```python
from sklearn import svm
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_scaled,y,test_size = 0.2,random_state = 1 , stratify = y)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_scaled,y,test_size = 0.2,random_state = 1 , stratify = y)

clf = svm.SVC(kernel='linear')
clf.fit(X_train,y_train)
y_predict = clf.predict(X_test)

print("Accuracy : ",metrics.accuracy_score(y_test,y_predict))
print("Precision : ",metrics.precision_score(y_test,y_predict))
print("Recall : ",metrics.recall_score(y_test,y_predict))

print("Confusion Matrix : ")
#fmt = format
mat = metrics.confusion_matrix(y_test,y_predict)
sns.heatmap(mat,square=True,annot=True,fmt='d',cbar=False,
            xticklabels=cancer.target_names,
            yticklabels=cancer.target_names)

plt.xlabel("Predicted")
plt.ylabel("True Labels")
plt.show()
```

# Polynomial Kernel

```python
X_train,X_test,y_train,y_test = train_test_split(X_scaled,y,test_size = 0.2,random_state = 1 , stratify = y)
```

```python
# Usually simple problems don't require higher degree polynomials
# more degree require more computational resources
clf = svm.SVC(kernel='poly',degree=2,gamma=100)
clf.fit(X_train,y_train)
y_predict = clf.predict(X_test)

print("Accuracy : ",metrics.accuracy_score(y_test,y_predict))
print("Precision : ",metrics.precision_score(y_test,y_predict))
print("Recall : ",metrics.recall_score(y_test,y_predict))

print("Confusion Matrix : ")
#fmt = format
mat = metrics.confusion_matrix(y_test,y_predict)
sns.heatmap(mat,square=True,annot=True,fmt='d',cbar=False,
            xticklabels=cancer.target_names,
            yticklabels=cancer.target_names)

plt.xlabel("Predicted")
plt.ylabel("True Labels")
plt.show()
```

# Sigmoid Kernel

```python
clf = svm.SVC(kernel='sigmoid',gamma=200,C=0.2)
clf.fit(X_train,y_train)
y_predict = clf.predict(X_test)

print("Accuracy : ",metrics.accuracy_score(y_test,y_predict))
print("Precision : ",metrics.precision_score(y_test,y_predict))
print("Recall : ",metrics.recall_score(y_test,y_predict))

print("Confusion Matrix : ")
#fmt = format
mat = metrics.confusion_matrix(y_test,y_predict)
sns.heatmap(mat,square=True,annot=True,fmt='d',cbar=False,
            xticklabels=cancer.target_names,
            yticklabels=cancer.target_names)

plt.xlabel("Predicted")
plt.ylabel("True Labels")
plt.show()
```
